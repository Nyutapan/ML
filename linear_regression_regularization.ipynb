1. Один из способов вычислить значения параметров модели является метод наименьших квадратов (МНК), который минимизирует среднеквадратичную ошибку между реальным значением зависимой переменной и прогнозом, выданным моделью:


$\large \begin{array}{rcl}\mathcal{L}\left(X, \vec{y}, \vec{w} \right) &=& \frac{1}{2n} \sum_{i=1}^n \left(y_i - \vec{w}^T \vec{x}_i\right)^2 \\ &=& \frac{1}{2n} \left\| \vec{y} - X \vec{w} \right\|_2^2 \\ &=& \frac{1}{2n} \left(\vec{y} - X \vec{w}\right)^T \left(\vec{y} - X \vec{w}\right) \end{array}$


Для решения данной оптимизационной задачи необходимо вычислить производные по параметрам модели, приравнять их к нулю и решить полученные уравнения относительно $\vec w$

$\large \begin{array}{rcl} \frac{\partial \mathcal{L}}{\partial \vec{w}} &=& \frac{\partial}{\partial \vec{w}} \frac{1}{2n} \left( \vec{y}^T \vec{y} -2\vec{y}^T X \vec{w} + \vec{w}^T X^T X \vec{w}\right) \\ &=& \frac{1}{2n} \left(-2 X^T \vec{y} + 2X^T X \vec{w}\right) \end{array}$
$\large \begin{array}{rcl} \frac{\partial \mathcal{L}}{\partial \vec{w}} = 0 &\Leftrightarrow& \frac{1}{2n} \left(-2 X^T \vec{y} + 2X^T X \vec{w}\right) = 0 \\ &\Leftrightarrow& -X^T \vec{y} + X^T X \vec{w} = 0 \\ &\Leftrightarrow& X^T X \vec{w} = X^T \vec{y} \\ &\Leftrightarrow& \vec{w} = \left(X^T X\right)^{-1} X^T \vec{y} \end{array}$



2. L2-регуляризации (Ridge).


$\large \begin{array}{rcl} \vec{w} &=& \left(X^T X + \lambda E\right)^{-1} X^T \vec{y} \end{array}$


Такая регрессия называется гребневой регрессией (ridge regression). А гребнем является диагональная матрица, которую мы прибавляем к матрице $X^T X$, в результате получается гарантированно регулярная матрица.

Для L1 (Lasso) аналитического решения нет — оптимизация выполняется численно.
Регуляризация добавляет штраф за большие коэффициенты, предотвращая переобучение.

В общем виде задача обучения с регуляризацией сводится к минимизации следующей функции, где к ошибке модели добавляется штраф за сложность весов:

$$\min_{w} \sum_{i=1}^{N} L(f(x_i, w), y_i) + \lambda R(w)$$

Где:
* $\sum L(...)$ — ошибка модели (например, MSE).
* $\lambda$ (или $\alpha$) — коэффициент регуляризации (гиперпараметр).
* $R(w)$ — функция регуляризации.

**Различия L1 и L2:**

1.  **L2 (Ridge):** $R(w) = \|w\|^2_2 = \sum w_i^2$.
    * Штрафует за квадрат величины весов.
    * Делает веса маленькими, но редко зануляет их полностью.
    * Имеет аналитическое решение: $w = (X^T X + \lambda I)^{-1} X^T y$.

2.  **L1 (Lasso):** $R(w) = \|w\|_1 = \sum |w_i|$.
    * Штрафует за сумму модулей весов.
    * **Не имеет** аналитического решения в замкнутом виде (требуются итеративные методы, например, координатный спуск).


3. Почему L1 часто используется для отбора признаков?

Основная причина, по которой L1-регуляризация (Lasso) используется для отбора признаков, заключается в её способности занулять веса (она создает разреженные (sparse) модели, где большинство неважных признаков имеют вес 0).

Когда вес признака становится равен 0, этот признак фактически исключается из уравнения регрессии.
Это приводит к автоматическому отбору признаков.

Формула штрафа для L1 выглядит так:

$$R(w) = \|w\|_1 = \sum_{i=1}^{d} |w_i|$$

Геометрически область ограничения для L1-нормы представляет собой ромб (в 2D) или многогранник с острыми вершинами на осях координат.

Когда линии уровня функции потерь (эллипсы MSE) касаются этой области ограничения, точка касания с высокой вероятностью приходится именно на **вершину** (на ось). В этой точке значение соответствующего веса $w_i$ становится строго равным **нулю**.

Это позволяет использовать Lasso как встроенный метод **отбора признаков** (Feature Selection): признаки с нулевыми весами исключаются из модели.

4. Можно преобразовать входные признаки в полиномиальные или другие нелинейные комбинации с помощью PolynomialFeatures.
Тогда даже линейная модель сможет аппроксимировать нелинейные зависимости. Также использование явных математических функций, таких как log(), exp(), sin(), cos(), sqrt(), — это классический и очень эффективный способ feature engineering (конструирования признаков).

### 4. Models implementation — Linear regression

Реализация собственного класса линейной регрессии, поддерживающий разные методы оптимизации:
1.  **Аналитическое решение (Normal Equation):** $\vec{w} = (X^T X)^{-1} X^T \vec{y}$. Дает точный результат мгновенно на малых данных, но требует вычисления обратной матрицы.
2.  **Градиентный спуск (GD):** Итеративное обновление весов.
3.  **Стохастический градиентный спуск (SGD):** Обновление весов по одному случайному объекту.

Также реализую метрику **$R^2$ (Coefficient of Determination)** для оценки качества модели.

```python
np.random.seed(42)
```

```python
class MyLinearRegression:
    def __init__(self, lr=0.01, n_iters=1000, method="gd"):
        self.lr = lr
        self.n_iters = n_iters
        self.method = method
        self.w = None
        self.b = None

    def predict(self, X):
        return X.dot(self.w) + self.b

    def fit_analytic(self, X, y):
        XTX = X.T.dot(X)
        XTy = X.T.dot(y)
        self.w = np.linalg.inv(XTX).dot(XTy)
        self.b = y.mean() - X.dot(self.w).mean()

    def fit_gd(self, X, y):
        n_samples, n_features = X.shape

        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            y_pred = X.dot(self.w) + self.b
            error = y_pred - y

            dw = (2/n_samples) * X.T.dot(error)
            db = (2/n_samples) * np.sum(error)

            self.w -= self.lr * dw
            self.b -= self.lr * db

    def fit_sgd(self, X, y, random_state=42):
        np.random.seed(random_state)

        n_samples, n_features = X.shape

        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            indices = np.arange(n_samples)
            np.random.shuffle(indices)

            for i in indices:
                xi = X[i]
                yi = y[i]

                y_pred = np.dot(xi, self.w) + self.b
                error = y_pred - yi

                dw = 2 * xi * error
                db = 2 * error

                self.w -= self.lr * dw
                self.b -= self.lr * db

    def fit(self, X, y):
        if self.method == "analytic":
            self.fit_analytic(X, y)
        elif self.method == "gd":
            self.fit_gd(X, y)
        elif self.method == "sgd":
            self.fit_sgd(X, y)
        else:
            raise ValueError("Unknown method")
```

Для оценки качества модели использую **Коэффициент детерминации ($R^2$)**. Эта метрика показывает, какую долю дисперсии целевой переменной ($y$) смогла объяснить модель по сравнению с базовой моделью, которая всегда предсказывает среднее значение.

$R^2$ рассчитывается по формуле:

$$R^2 = 1 - \frac{\sum_{i=1}^{N} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}$$

* **Числитель ($\sum (y_i - \hat{y}_i)^2$): $\text{SSE}$ (Sum of Squares of Errors) / $\text{SS}_{\text{res}}$**
    * Сумма квадратов ошибок вашей модели. Это то, насколько плохо предсказывает ваша модель.
* **Знаменатель ($\sum (y_i - \bar{y})^2$): $\text{SST}$ (Total Sum of Squares) / $\text{SS}_{\text{tot}}$**
    * Общая сумма квадратов. Это ошибка базовой модели, которая всегда предсказывает среднее ($\bar{y}$).

### Вывод:

* **$R^2 \approx 1$**: Модель идеальна (почти нет необъясненной дисперсии).
* **$R^2 \approx 0$**: Модель работает не лучше, чем простое среднее.
* **$R^2 < 0$**: Модель работает хуже, чем базовая модель (признак очень сильного переобучения).

```python
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - ss_res / ss_tot
```

### 5. Regularized models implementation — Ridge, Lasso, ElasticNet

Чтобы предотвратить переобучение добавляю регуляризацию (штраф за большие веса) в функцию потерь.

* **Ridge (L2):** Добавляет штраф $\alpha \sum w_i^2$. Веса стремятся к нулю, но не обнуляются. Хорошо работает, когда все признаки важны.
* **Lasso (L1):** Добавляет штраф $\alpha \sum |w_i|$. Может занулять веса, выполняя **отбор признаков**.
* **ElasticNet:** Комбинация L1 и L2.

Реализую эти модели вручную, модифицируя шаг обновления весов в градиентном спуске, и сравниваю их с реализациями из `sklearn`.

```python
class MyRidgeRegression:
    def __init__(self, lr=0.01, n_iters=1000, l2=0.1):
        self.lr = lr
        self.n_iters = n_iters
        self.l2 = l2

    def fit(self, X, y):
        n, d = X.shape
        self.w = np.zeros(d)
        self.b = 0

        for _ in range(self.n_iters):
            y_pred = X @ self.w + self.b

            dw = (-2/n) * (X.T @ (y - y_pred)) + 2 * self.l2 * self.w
            db = (-2/n) * np.sum(y - y_pred)

            self.w -= self.lr * dw
            self.b -= self.lr * db

    def predict(self, X):
        return X @ self.w + self.b
```

```python
class MyLassoRegression:
    def __init__(self, lr=0.01, n_iters=1000, l1=0.1):
        self.lr = lr
        self.n_iters = n_iters
        self.l1 = l1

    def fit(self, X, y):
        n, d = X.shape
        self.w = np.zeros(d)
        self.b = 0

        for _ in range(self.n_iters):
            y_pred = X @ self.w + self.b

            grad_w = (-2/n) * (X.T @ (y - y_pred)) + self.l1 * np.sign(self.w)
            grad_b = (-2/n) * np.sum(y - y_pred)

            self.w -= self.lr * grad_w
            self.b -= self.lr * grad_b

    def predict(self, X):
        return X @ self.w + self.b
```

```python
class MyElasticNet:
    def __init__(self, lr=0.01, n_iters=1000, l1=0.1, l2=0.1):
        self.lr = lr
        self.n_iters = n_iters
        self.l1 = l1
        self.l2 = l2

    def fit(self, X, y):
        n, d = X.shape
        self.w = np.zeros(d)
        self.b = 0

        for _ in range(self.n_iters):
            y_pred = X @ self.w + self.b

            grad_w = (-2/n) * (X.T @ (y - y_pred)) + \
                     self.l1 * np.sign(self.w) + \
                     2 * self.l2 * self.w

            grad_b = (-2/n) * np.sum(y - y_pred)

            self.w -= self.lr * grad_w
            self.b -= self.lr * grad_b

    def predict(self, X):
        return X @ self.w + self.b
```

```python
def evaluate(model, model_name, X_train, y_train, X_test, y_test,
             result_MAE, result_RMSE, result_R2):

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    mae_train = mean_absolute_error(y_train, y_train_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)

    result_MAE.loc[len(result_MAE)] = [model_name, mae_train, mae_test]
    result_RMSE.loc[len(result_RMSE)] = [model_name, rmse_train, rmse_test]
    result_R2.loc[len(result_R2)] = [model_name, r2_train, r2_test]

    return result_MAE, result_RMSE, result_R2
my_models = {
    "my_ridge": MyRidgeRegression(lr=0.01, n_iters=2000, l2=0.1),
    "my_lasso": MyLassoRegression(lr=0.01, n_iters=2000, l1=0.1),
    "my_elnet": MyElasticNet(lr=0.01, n_iters=2000, l1=0.1, l2=0.1)
}

for name, model in my_models.items():
    model.fit(X_train.values, y_train.values)
    result_MAE, result_RMSE, result_R2 = evaluate(
        model, name,
        X_train.values, y_train.values,
        X_test.values, y_test.values,
        result_MAE, result_RMSE, result_R2
    )
sk_models = {
    "sk_ridge": Ridge(alpha=1.0),
    "sk_lasso": Lasso(alpha=0.001),
    "sk_elnet": ElasticNet(alpha=0.001, l1_ratio=0.5)
}

for name, model in sk_models.items():
    model.fit(X_train, y_train)
    result_MAE, result_RMSE, result_R2 = evaluate(
        model, name,
        X_train, y_train,
        X_test, y_test,
        result_MAE, result_RMSE, result_R2
    )
```

### 6. Feature normalization

Нормализация обязательна:

Модели с градиентным спуском (линейная/логистическая регрессия, нейросети)

Модели, основанные на расстояниях (KNN, K-Means, DBSCAN)

SVM с RBF-ядром

Не обязательна:

Деревья решений, Random Forest, XGBoost

Наивный Байес

Min_max:

$$
x_{\text{scaled}} = \frac{x - \min_{\text{original}}}{\max_{\text{original}} - \min_{\text{original}}}
$$

Standart:

$$
x_{\text{scaled}} = \frac{x - \mu}{\sigma}
$$

Линейные модели, обучаемые градиентным спуском, чувствительны к масштабу данных. Если один признак измеряется в единицах (количество комнат), а другой в тысячах (площадь), градиент будет сходиться очень медленно.

Реализую и сравню два метода:
1.  **MinMaxScaler:** $x' = \frac{x - \min}{\max - \min}$. Приводит данные к диапазону [0, 1].
2.  **StandardScaler:** $x' = \frac{x - \mu}{\sigma}$. Приводит данные к нормальному распределению с $\mu=0$ и $\sigma=1$.

Ожидается, что на нормализованных данных методы градиентного спуска будут сходиться быстрее и стабильнее.

```python
class MyMinMaxScaler:
    def fit(self, X):
        X = np.array(X)
        self.min_ = X.min(axis=0)
        self.max_ = X.max(axis=0)
        return self

    def transform(self, X):
        X = np.array(X)
        return (X - self.min_) / (self.max_ - self.min_)

    def fit_transform(self, X):
        return self.fit(X).transform(X)
```

```python
class MyStandardScaler:
    def fit(self, X):
        X = np.array(X)
        self.mean_ = X.mean(axis=0)
        self.std_ = X.std(axis=0)
        return self

    def transform(self, X):
        X = np.array(X)
        return (X - self.mean_) / self.std_

    def fit_transform(self, X):
        return self.fit(X).transform(X)
```

```python
def add_metrics(model_name, y_train, y_train_pred, y_test, y_test_pred,
                result_MAE, result_RMSE, result_R2):

    result_MAE.loc[len(result_MAE)] = [
        model_name,
        mean_absolute_error(y_train, y_train_pred),
        mean_absolute_error(y_test, y_test_pred)
    ]

    result_RMSE.loc[len(result_RMSE)] = [
        model_name,
        np.sqrt(mean_squared_error(y_train, y_train_pred)),
        np.sqrt(mean_squared_error(y_test, y_test_pred))
    ]

    result_R2.loc[len(result_R2)] = [
        model_name,
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)
    ]
```

```python
def train_and_evaluate_models(
        X_train, X_test, y_train, y_test,
        result_MAE, result_RMSE, result_R2,
        ridge_alpha=1.0, lasso_alpha=1.0,
        elastic_alpha=1.0, elastic_l1_ratio=0.5,
        scaler='none',
        postfix=''
    ):

    # нормализация
    if scaler == 'minmax':
        sc = MinMaxScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)

    elif scaler == 'standard':
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)

    models = {
        "Linear": LinearRegression(),
        "Ridge": Ridge(alpha=ridge_alpha),
        "Lasso": Lasso(alpha=lasso_alpha),
        "ElasticNet": ElasticNet(alpha=elastic_alpha,
                                 l1_ratio=elastic_l1_ratio)
    }

    for name, model in models.items():
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred  = model.predict(X_test)

        add_metrics(
            name + "_" + postfix,
            y_train, y_train_pred,
            y_test, y_test_pred,
            result_MAE, result_RMSE, result_R2
        )

    return result_MAE, result_RMSE, result_R2
```

Данные после собственного MinMaxScaler и StandardScaler полностью совпадают с результатами sklearn. Это означает, что обе реализации работают корректно.

Для StandardScaler разница = 0,

### 11. Addition task

Для повышения качества модели можно применить несколько техник:

1.  **Логарифмирование целевой переменной:** Распределение цен часто имеет "длинный хвост" (много дешевых и мало очень дорогих квартир). Преобразование $y' = \log(1+y)$ делает распределение более нормальным, что улучшает работу линейной модели. При расчете метрик делаю обратное преобразование `exp`.*
2.  **Удаление выбросов из Train:** Удаляю экстремальные значения только из обучающей выборки. Тестовую выборку не трогаю, чтобы честно оценить работу модели на реальных данных.
3.  **Mini-Batch Gradient Descent:** Реализация компромисса между GD и SGD для более стабильного и быстрого обучения.

Результаты покажут, как техники справились с неочищенной тестовой выборкой

```python
class LinearBatchGD:
    def __init__(self, lr=1, n_iters=6000):
        self.lr = lr
        self.n_iters = n_iters
    def fit(self, X, y):
        m, n = X.shape
        X = np.c_[np.ones(m), X]
        self.w = np.zeros(n+1)
        y = y.values if isinstance(y, pd.Series) else y

        for _ in range(self.n_iters):
            y_pred = X @ self.w
            grad = (1/m) * (X.T @ (y_pred - y))
            self.w -= self.lr * grad
    def predict(self, X):
        X = np.c_[np.ones(X.shape[0]), X]
        return X @ self.w

class LinearMiniBatchGD:
    def __init__(self, lr=0.001, n_iters=6000, batch_size=32):
        self.lr = lr
        self.n_iters = n_iters
        self.batch_size = batch_size

    def fit(self, X, y):
        m, n = X.shape
        X = np.c_[np.ones(m), X]
        self.w = np.zeros(n+1)
        y = y.values if isinstance(y, pd.Series) else y

        for _ in range(self.n_iters):
            indices = np.random.permutation(m)
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            for i in range(0, m, self.batch_size):
                X_b = X_shuffled[i:i+self.batch_size]
                y_b = y_shuffled[i:i+self.batch_size]

                y_pred = X_b @ self.w
                grad = (1/len(X_b)) * (X_b.T @ (y_pred - y_b))
                self.w -= self.lr * grad

    def predict(self, X):
        X = np.c_[np.ones(X.shape[0]), X]
        return X @ self.w

# Масштабирование
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# y_train, y_test остаются не масштабированными

# Linear_log
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)

model_log = LinearRegression()
model_log.fit(X_train_scaled, y_train_log)

# Возврат из лог пространства
y_train_pred = np.expm1(model_log.predict(X_train_scaled))
y_test_pred = np.expm1(model_log.predict(X_test_scaled))

# Расчет метрик
log_r2 = (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred))

result_R2.loc[len(result_R2)] = ["Linear_log", log_r2[0], log_r2[1]]

upper = y_train.quantile(0.99)
lower = y_train.quantile(0.01)
mask = (y_train >= lower) & (y_train <= upper)

# Создание очищенных выборок
X_train_clean = X_train[mask]
y_train_clean = y_train[mask]

# Повторное масштабирование на очищенных данных
scaler_clean = StandardScaler()
X_train_clean_scaled = scaler_clean.fit_transform(X_train_clean)
X_test_clean_scaled = scaler_clean.transform(X_test)

model_clean = LinearRegression()
model_clean.fit(X_train_clean_scaled, y_train_clean)

y_train_pred = model_clean.predict(X_train_clean_scaled)
y_test_pred = model_clean.predict(X_test_clean_scaled)

# Расчет метрик
clean_r2 = (r2_score(y_train_clean, y_train_pred), r2_score(y_test, y_test_pred))

result_R2.loc[len(result_R2)] = ["Linear_no_outliers", clean_r2[0], clean_r2[1]]

# Batch GD
model_bgd = LinearBatchGD(lr=0.001, n_iters=6000)
model_bgd.fit(X_train_scaled, y_train)

y_train_pred = model_bgd.predict(X_train_scaled)
y_test_pred = model_bgd.predict(X_test_scaled)

bgd_r2 = (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred))
result_R2.loc[len(result_R2)] = ["Linear_BGD_STD", bgd_r2[0], bgd_r2[1]]

# Mini-Batch GD
model_mbgd = LinearMiniBatchGD(lr=0.001, n_iters=6000, batch_size=32)
model_mbgd.fit(X_train_scaled, y_train)

y_train_pred = model_mbgd.predict(X_train_scaled)
y_test_pred = model_mbgd.predict(X_test_scaled)

mbgd_r2 = (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred))
result_R2.loc[len(result_R2)] = ["Linear_MBGD_STD", mbgd_r2[0], mbgd_r2[1]]
```
